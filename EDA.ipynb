#### Bivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr

def numeric_bivariate_analysis(df, col_x, col_y):
    # Drop NA for clean calculation
    data = df[[col_x, col_y]].dropna()
    
    # Calculate correlations
    pearson_corr, p_p = pearsonr(data[col_x], data[col_y])
    spearman_corr, p_s = spearmanr(data[col_x], data[col_y])
    
    print(f" Pearson correlation: {pearson_corr:.3f} (p={p_p:.3g})")
    print(f" Spearman correlation: {spearman_corr:.3f} (p={p_s:.3g})")
    
    # Scatter plot
    plt.figure(figsize=(7, 5))
    sns.scatterplot(data=data, x=col_x, y=col_y, alpha=0.6, edgecolor='w')
    plt.title(f"Scatter Plot: {col_x} vs {col_y}")
    plt.xlabel(col_x)
    plt.ylabel(col_y)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway, kruskal

def categorical_price_analysis(df, cat_col, num_col):
    # Drop NA
    data = df[[cat_col, num_col]].dropna()
    
    print(f" Summary stats for {num_col} by {cat_col}:")
    print(data.groupby(cat_col)[num_col].agg(['count', 'mean', 'median', 'min', 'max']).sort_values('mean', ascending=False))
    print("\n")
    
    # Boxplot
    plt.figure(figsize=(10,10))
    sns.boxplot(data=data, x=cat_col, y=num_col)
    plt.xticks(rotation=45)
    plt.title(f"{num_col} Distribution by {cat_col} (Boxplot)")
    plt.show()
    
    
```
